```python
import pandas as pd
```

```python
date = "Mei 2022"

df_sentiment = pd.read_csv(f"../CSV/{date}/sentiment_dataset_labelled.csv")
try:
    df_sentiment = df_sentiment.drop("Unnamed: 0", axis = 1)
except:
    None
df_sentiment
```

```python
df = pd.read_csv(f"../../Training/data/dataset_type.csv")

df["BAD NEWS"] = 0
df["NEUTRAL NEWS"] = 0
df["GOOD NEWS"] = 0

df
```

```python
count = 0
double_count = 0
for i in range(len(df_sentiment)):
    for j in range(len(df)):
        if df["MONTH"][j] != 5 or df["YEAR"][j] != 2022:
            continue

        if df_sentiment["Brand"][i].lower() == df["BRAND"][j].lower():
            if df["TYPE"][j].lower() in df_sentiment["Title"][i].lower():
                double_count += 1
                
                if df_sentiment["Label"][i] == -1:
                    df["BAD NEWS"][j] += 1
                elif df_sentiment["Label"][i] == 0:
                    df["NEUTRAL NEWS"][j] += 1
                else:
                    df["GOOD NEWS"][j] += 1
            
            if df_sentiment["Label"][i] == -1:
                df["BAD NEWS"][j] += 1
            elif df_sentiment["Label"][i] == 0:
                df["NEUTRAL NEWS"][j] += 1
            else:
                df["GOOD NEWS"][j] += 1
            # print("================================================")
            # print("------ Sentiment ------")
            # print(df_sentiment["Title"][i])
            # print(df_sentiment["Brand"][i])
            # print(df_sentiment["Label"][i])
                
            # print("------  Dataset  ------")
            # print(df["BRAND"][j])
            # print(df["TYPE"][j])
            # print(str(df["YEAR"][j]) + " - " + str(df["MONTH"][j]))
            # print("================================================")
            count += 1
```
