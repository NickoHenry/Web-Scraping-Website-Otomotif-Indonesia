```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
def brand_valid(title, brand):
    
    if brand.lower() in title.lower():
        return True
    elif "VW".lower() in title.lower() and brand.lower() == "volkswagen":
        return True
    elif "Benz".lower() in title.lower() and brand.lower() == "mercedes":
        return True

    return False
```

```python
brands = ["Audi", "BMW", "Chery", "Chevrolet", "Chrysler", "Daihatsu", "Datsun", "Ford", "Foton", "Great Wall", "Hino", "Honda", "Hummer", "Hyundai", "Isuzu", "Jaguar", "Jeep", "Kia", "Land Rover", "Lexus", "Mazda", "Mercedes", "Mitsubishi", "Nissan", "Opel", "Peugeot", "Proton", "Renault", "SSangyong", "Subaru", "Suzuki", "Toyota", "UD Trucks", "Volkswagen", "Volvo", "Wuling", "Yamaha"]
titles_list = []
brands_list = []
count = 0
url = "https://autonetmagz.com/search/"
month = "May" #bulan yang ingin dicari
year = 2022

for brand in brands:
    stop = False
    page = 1
    brand_count = 0
    
    while True:
        driver.get(url + brand + "/page/" + str(page))
        content = driver.page_source
        soup = BeautifulSoup(content)
        page += 1

        for news in soup.findAll('article', attrs={'class':'autonetmagz__blog--grid'}):
            title = news.find('h3', attrs={'class':'entry-title'}).text
            date = news.find('time', attrs={'class':'updated'}).text

            if brand_valid(title, brand) and month in date and str(year) in date:
                titles_list.append(title)
                brands_list.append(brand)
                    
                
                count += 1
                brand_count += 1

            if str(year) not in date and "lalu" not in date:
                stop = True
                break

        if stop == True:
            break
print("------------------------------------\nTotal news this month: " + str(count))
```

```python
date = "Mei 2022"

df = pd.DataFrame({'Title':titles_list, 'Brand':brands_list})
df = df.sort_values(by=['Brand'])
df.to_csv(f"../../CSV/{date}/autonetmagz.csv", index=False)
```
